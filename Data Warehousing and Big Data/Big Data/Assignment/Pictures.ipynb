```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("healthcare_dataset.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 4))
ax = sns.countplot(data=df, x='Gender', palette='pastel')

# Add title and labels
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add actual numbers on the bars
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1, int(height), ha='center', va='bottom')

plt.show()
```

```python
plt.figure(figsize=(8, 4))
ax = sns.countplot(data=df, x='Medical Condition', palette='pastel')

# Add title and labels
plt.title('Distribution of Various Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Count')

# Add actual numbers on the bars
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1, int(height), ha='center', va='bottom')

# Rotate x-axis labels if necessary for better readability
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
plt.figure(figsize=(8, 4))
ax = sns.countplot(data=df, x='Admission Type', hue='Gender', palette='pastel')

# Add title and labels
plt.title('Admission Types Across Different Genders')
plt.xlabel('Admission Type')
plt.ylabel('Count')

# Add actual numbers on the bars
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1, int(height), ha='center', va='bottom')

# Rotate x-axis labels if necessary for better readability
plt.xticks(rotation=45, ha='right')

plt.legend(title='Gender')
plt.show()
```

```python
# Aggregate the number of patients per doctor
doctor_counts = df['Doctor'].value_counts()

# Get the top 10 doctors
top_10_doctors = doctor_counts.head(10)

# Create a DataFrame for easy plotting
top_10_df = top_10_doctors.reset_index()
top_10_df.columns = ['Doctor', 'Number of Patients']

# Plotting
plt.figure(figsize=(8, 4))
ax = sns.barplot(data=top_10_df, x='Number of Patients', y='Doctor', palette='pastel')

# Add title and labels
plt.title('Top 10 Doctors by Number of Patients')
plt.xlabel('Number of Patients')
plt.ylabel('Doctor')

# Add actual numbers on the bars
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 0.1, p.get_y() + p.get_height() / 2., int(width), ha='left', va='center')

plt.show()
```
